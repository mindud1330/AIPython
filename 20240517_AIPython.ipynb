# Graph
## Matplotlib
### plot

```python
import matplotlib.pyplot as plt
import pandas as pd

df=pd.read_excel('data1.xlsx')

plt.plot(df['kor'], marker='o', label='korean')
plt.plot(df['math'], marker='^',label='math')
plt.legend()
plt.show()
```

### scatter

```python
import numpy as np

x = np.arange(1, 11)
y = np.random.randint(1, 30, 10)

plt.scatter(x, y)
plt.show()
```

### bar

```python
x = ['A', 'B', 'C']
y = [100, 150, 120]

plt.bar(x, y)
plt.title('Test Graph')
plt.xlabel('Group')
plt.ylabel('Score')

plt.show()
```

### histogram

```python
x= [1, 1, 1, 2, 3, 2, 4]

plt.hist(x, bins=4, color='green', align='mid')
plt.show()
```

### subplots

```python
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

fig, axs = plt.subplots(2,2)

axs[0,0].plot(x,y)
axs[0,1].scatter(x,y)
axs[1,0].bar(x,y)
axs[1,1].barh(x,y)
fig.suptitle('Drawing Graph')
plt.show()
```

## Seaborn
### scatterplot

```python
import seaborn as sns

# iris_data=pd.read_csv('iris.csv')
iris_data = sns.load_dataset("iris")

# scatter plot
sns.scatterplot(x = 'petal_length', y = 'petal_width' ,hue = 'species',
                data = iris_data)
plt.title('Scatter Plot by seaborn', fontsize=20)
plt.show()
```

### countplot

```python
# iris_data=pd.read_csv('iris.csv')
iris_data = sns.load_dataset("iris")

sns.countplot(x='species', hue='species', data=iris_data)
plt.show()
```

### barplot

```python
# iris_data=pd.read_csv('iris.csv')
iris_data = sns.load_dataset("iris")

sns.barplot(x='species',y='petal_length', hue='species',
            data=iris_data, errorbar='sd')
plt.show()
```

### pairplot

```python
# iris_data=pd.read_csv('iris.csv')
iris_data = sns.load_dataset("iris")

sns.pairplot(iris_data,hue='species')
plt.show()
```

### regplot

```python
# iris_data=pd.read_csv('iris.csv')
iris_data = sns.load_dataset("iris")

sns.regplot(x='petal_length',y='petal_width', data=iris_data)
plt.show()
```

### heatmap

```python
# iris_data=pd.read_csv('iris.csv')
iris_data = sns.load_dataset("iris")
numeric_iris_data = iris_data.select_dtypes(include=['float64', 'int64'])

correlation_matrix = numeric_iris_data.corr()

sns.heatmap(data=correlation_matrix, annot=True, fmt='.2f', cbar=False)
plt.show()
```
